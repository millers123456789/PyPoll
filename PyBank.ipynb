```python
#The value_counts method counts unique values in a column. In our case we want to count the months
count = pybank_pd["Date"].value_counts()
count
#As can be seen each date is different so it gives each one count.
#It does give the toal amount of months 86
```

```python
#We want the toal net amount of "Profit/Losses" over the entire period
#This is where we run our sum function
total = pybank_pd["Profit/Losses"].sum()
total
```

```python
#Now we have to calculate the average change between months. 
#First calculate the change between months
diff = pybank_pd["Profit/Losses"].diff()
diff
```

```python
#Find the mean of differences by taking the average of the data frame diff_df
average = diff_df.mean()
average
```
